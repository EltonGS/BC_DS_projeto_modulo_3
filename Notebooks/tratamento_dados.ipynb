## Informações sobre a base de dados

Ao verificar a página de onde os dados foram extraídos, [Brasil.IO](https://brasil.io/dataset/covid19/caso_full/), temos a informaçãos que alguns dados podem ser repetidos no dataset quando os dados não foram divulgados pela **SES - Secretarias Estaduais de Saúde**.

No dataset temos as colunas **is_last** e **is_repeated** que identificam quando um dado é a **última atualização disponível** e quando o **dado é repetido**, respectivamente.

```python
# importando as bibliotecas que serão usadas
import pandas as pd
import numpy as np
```

```python
# caminho de onde estão os dados utilizados
url = '/content/drive/MyDrive/Colab Notebooks/Bootcamp Data Science/Módulo 3/Arquivos/caso_full.zip'

# lendo o arquivo com os dados que estão comprimidos no padrão zip
dados_totais = pd.read_csv(url, compression='zip')
```

```python
# visualizando os dados
dados_totais.head(3)
```

### É possível perceber uma linha abaixo dos dados da cidade de Rio Branco que parece repetir o dado da linha acima, ao repararmos na coluna place_type vemos que essa segunda linha se refere ao estado onde a cidade está. Para a cidade de Rio Branco temos os dados do estado do Acre.

### Para esse trabalho focaremos na cidade de Santo André no estado de São Paulo, onde resido.

```python
# verificando os dados que são da cidade (city) santo andré E do estado (state) SP
dados_sa = dados_totais.query("city == 'Santo André' & state == 'SP'")
dados_sa
```

Nos dados mostrados acima temos 470 linhas para a cidade Santo André em SP, e temos também o código IBGE da cidade: **3547809.0**

Assim podemos verificar se temos todos os dados referente a essa cidade.

```python
# extração dos dados pelo código da cidade
dados_sa2 = dados_totais.query("city_ibge_code == 3547809.0")
dados_sa2
```

Os dados extraídos das duas maneiras são os mesmos. Continuaremos com os dados do dataframe **dados_sa**.

```python
# verificação se há alguma dado NaN em nosso dataframe
dados_sa.isna().value_counts()
```

```python
# redefinindo o index para facilitar o trabalho com os dados
dados_sa = dados_sa.reset_index()

# dados com o index resetado
dados_sa.head(3)
```

```python
# verificando se temos alguma informação True na coluna is_last
dados_sa['is_last'].value_counts()
```

```python
# verificando onde está o dado True identificado acima
dados_sa.query("is_last == True")
```

```python
# verificando os dados da linha 468 do index
dados_sa.loc[468]
```

Podemos ver na informação acima que a **última atualização disponível** para **Santo André / SP** é da data **27/06/2021**.

Agora vamos verificar se em nosso dataframe temos algum dado repetido.

```python
# verificando se temos alguma informação True na coluna is_repeated
dados_sa['is_repeated'].value_counts()
```

```python
# verificando quais são os dados repetidos que temos no datafarme
dados_repetidos = dados_sa.query("is_repeated == True")
dados_repetidos
```

No dataframe acima (**dados_repetidos**) temos as datas em que os dados foram repetidos por não ter informação divulgada pelo boletime das Secretarias Estaduais de Saúde.

Temos um total de **28** datas.

```python
dados_sa.head(5)
```

```python
# renomeando as colunas para facilitar o trabalho com os dados
dados_sa.rename(columns={
    'city' : 'cidade',
    'city_ibge_code' : 'codigo_cidade',
    'date' : 'data',
    'epidemiological_week' : 'semana_epidem',
    'estimated_population' : 'populacao',
    'estimated_population_2019' : 'populacao_2019',
    'last_available_confirmed' : 'casos_confirmados',
    'last_available_confirmed_per_100k_inhabitants' : 'casos_por_100k_habitantes',
    'last_available_date' : 'data_atualizacao',
    'last_available_death_rate' : 'taxa_mortalidade',
    'last_available_deaths' : 'obitos',
    'state' : 'estado',
    'new_confirmed' : 'casos_novos',
    'new_deaths' : 'obitos_novos'}, inplace=True)
```

```python
# visualizando os novos nomes das colunas
dados_sa.head(3)
```

```python
# removendo as colunas que não interessam para o projeto
dados_sa.drop(['index', 'order_for_place', 'place_type'], axis=1, inplace=True)
```

```python
dados_sa.head(3)
```

Agora temos os dados para **Santo André / SP** tratados.

Foram identificadas as linhas com dados repetidos, se haviam dados NaN, o index foi resetados, as colunas foram renomeadas e as colunas que não interessam para essa análise foram removidas.

Para efeito de comparação com os dados de Santo André, também serão extraídos os dados do estado de São Paulo.


```python
# extraindo a informação para o estado de SP
dados_estadual_sp = dados_totais.query("state == 'SP' & place_type == 'state'")
```

```python
# visualizando os dados para o estado de SP
dados_estadual_sp
```

Temos um dataframe com **490 linhas** e **18 colunas**.

```python
# extraindo os dados pelo código do IBGE para comparar
dados_estadual_sp2 = dados_totais.query("city_ibge_code == 35.0")

# visualizando os dados extraídos
dados_estadual_sp2
```

A informação é igual a que foi extraída anteriormente. A partir daqui continuaremos com o dataframe **dados_estadual_sp**.

```python
# verificando se temos dados NaN em nosso dataframe
dados_estadual_sp.isna().value_counts()
```

Na coluna **city** temos dados NaN pois os dados não são referentes a nenhuma cidade mas sim ao estado de São Paulo.

Para tornar o trabalho com os dados mais intuitivo e simples, substituiremos a informação NaN pelo nome do estado São Paulo.

```python
# substituindo a informação NaN na coluna city pelo nome do estado São Paulo
dados_estadual_sp['city'].replace(np.nan, 'São Paulo', inplace=True)

# visualizando os dados com a alteração na coluna city
dados_estadual_sp.head(3)
```

```python
# verificando se há mais algum dados NaN no dataframe
dados_estadual_sp.isna().value_counts()
```

```python
# resetando o index para facilitar o trabalho com os dados
dados_estadual_sp = dados_estadual_sp.reset_index()

# visualizando os dados com o novo index
dados_estadual_sp.head(3)
```

```python
# verificando a coluna is_last se temos algum dado True
dados_estadual_sp.query("is_last == True")
```

```python
dados_estadual_sp.tail(3)
```

Podemos ver que a informação disponibilizada para o dia 28/06/2021 é a última disponibilizada para o estado de SP e é a última linha no dataframe.

```python
# verificando a coluna is_repeated se temos algum dado True
dados_repetidos_sp = dados_estadual_sp.query("is_repeated == True")
```

```python
# visualizando os dados com True na coluna is_repeated
dados_repetidos_sp
```

No dataframe acima (**dados_repetidos_sp**) temos as datas em que os dados foram repetidos por não ter informação divulgada pelo boletim das Secretarias Estaduais de Saúde.

Temos um total de **7** datas.

```python
# renomeando as colunas para facilitar o trabalho com os dados
dados_estadual_sp.rename(columns={
    'city' : 'estado',
    'city_ibge_code' : 'codigo_estado',
    'date' : 'data',
    'epidemiological_week' : 'semana_epidem',
    'estimated_population' : 'populacao',
    'estimated_population_2019' : 'populacao_2019',
    'last_available_confirmed' : 'casos_confirmados',
    'last_available_confirmed_per_100k_inhabitants' : 'casos_por_100k_habitantes',
    'last_available_date' : 'data_atualizacao',
    'last_available_death_rate' : 'taxa_mortalidade',
    'last_available_deaths' : 'obitos',
    'state' : 'estado',
    'new_confirmed' : 'casos_novos',
    'new_deaths' : 'obitos_novos'}, inplace=True)
```

```python
# visualizando as colunas renomeadas
dados_estadual_sp.head(3)
```

```python
# removendo as colunas que não interessam para o projeto
dados_estadual_sp.drop(['index', 'order_for_place', 'place_type'], axis=1, inplace=True)
```

```python
# visualizando o dataframe com as colunas que interessam para o projeto
dados_estadual_sp.head(3)
```

Agora temos os dados para o estado de **São Paulo** tratados.

Foram identificadas as linhas com dados repetidos, se haviam dados NaN, os dados NaN identificados foram substituídos por São Paulo, o index foi resetado, as colunas foram renomeadas e as colunas que não interessam para essa análise foram removidas.

Agora os dataframes que serão utilizados na análise explaratória serão exportados como arquivos CSV.

```python
# exportando o dataframe com os dados de Santo André como arquivo CSV
dados_sa.to_csv('dados_sa.csv', index=False)

# exportando o dataframe com os dados do estado de São Paulo como arquivo CSV
dados_estadual_sp.to_csv('dados_estadual_sp.csv', index=False)

# exportando o dataframe com os dados repetidos de Santo André como arquivo CSV
dados_repetidos.to_csv('dados_repetidos_sa.csv', index=False)

# exportando o dataframe com os dados repetidos do estado de SP como arquivo CSV
dados_repetidos_sp.to_csv('dados_repetidos_sp.csv', index=False)
```
